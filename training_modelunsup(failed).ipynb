```python
# Initial imports
import pandas as pd
import hvplot.pandas
from path import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
#import the csv file and convert it to a dataframe 

CleanedData_df = pd.read_csv(Path('./Resources/CleanedData.csv'))
CleanedData_df.head()
```

```python
#Remove Patient ID Column
CleanedData_df.drop(columns=["Patient Id"], inplace=True)

CleanedData_df.head()
```

```python
# Initializing model with K = 3 (since we already know there are three classes of cancer severity)
model = KMeans(n_clusters=3, random_state=5)
model
```

```python
# Fitting model
model.fit(CleanedData_df)
```
